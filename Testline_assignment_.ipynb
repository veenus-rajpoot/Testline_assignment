```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Data Exploration
# Load data from JSON endpoints
with open('current_quiz.json') as f:
    current_quiz = json.load(f)

with open('quiz_submission.json') as f:
    quiz_submission = json.load(f)

with open('historical_quiz.json') as f:
    historical_quiz = json.load(f)
```

```python
# Convert data into Pandas DataFrame
current_quiz_df = pd.json_normalize(current_quiz['quiz']['questions'])

# Extract quiz submission main data and response map
quiz_submission_main = pd.json_normalize(quiz_submission)


# Convert historical quiz data into DataFrame
historical_quiz_df = pd.DataFrame(historical_quiz)
```

```python
def calculate_accuracy(correct, total):
    return (correct / total) * 100 if total > 0 else 0
```

```python
current_accuracy = calculate_accuracy(
    quiz_submission_main['correct_answers'].iloc[0],
    quiz_submission_main['total_questions'].iloc[0]
)
```

```python
historical_accuracy = calculate_accuracy(
    historical_quiz_df['correct_answers'].sum(),
    historical_quiz_df['total_questions'].sum()
)
```

```python

# Identify weak areas
def weak_areas(df):
    topic_accuracy = df.groupby('topic')['correct_answers'].sum() / df.groupby('topic')['total_questions'].sum()
    return topic_accuracy.sort_values()
```

```python
# Flatten response maps and associate metadata
flattened_responses = []
for quiz in historical_quiz:
    quiz_id = quiz["quiz_id"]
    topic = quiz["quiz"]["topic"]
    user_id = quiz["user_id"]
    response_map = quiz["response_map"]

    for question_id, selected_option in response_map.items():
        flattened_responses.append({
            "quiz_id": quiz_id,
            "topic": topic,
            "user_id": user_id,
            "question_id": question_id,
            "selected_option": selected_option,
        })

# Convert to DataFrame
historical_responses_df = pd.DataFrame(flattened_responses)

# Example visualization of response trends by topic
response_summary = historical_responses_df.groupby("topic").size()
sns.barplot(x=response_summary.index, y=response_summary.values)
plt.title("Response Distribution by Topic")
plt.xlabel("Topics")
plt.ylabel("Number of Responses")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Calculate weak areas (using mock data for correct answers and total questions)
def compute_topic_accuracy(historical_data, topic_accuracies):
    for topic, group in historical_data.groupby("topic"):
        # Assuming topic_accuracies has mock correct and total question counts
        correct = topic_accuracies.get(topic, {}).get("correct_answers", 0)
        total = topic_accuracies.get(topic, {}).get("total_questions", 1)
        topic_accuracy = (correct / total) * 100
        print(f"Topic: {topic}, Accuracy: {topic_accuracy:.2f}%")

# Mock topic accuracy data
mock_topic_accuracies = {
    "Body Fluids and Circulation": {"correct_answers": 80, "total_questions": 100},
    # Add more topics as needed
}

compute_topic_accuracy(historical_responses_df, mock_topic_accuracies)
```

```python
# Step 2: Generate Insights
# Mock topic accuracy data
mock_topic_accuracies = {
    "Body Fluids and Circulation": {"correct_answers": 80, "total_questions": 100},
    # Add more topics as needed
}
```

```python
def compute_topic_accuracy(historical_data, topic_accuracies):
    topic_accuracy_results = []
    for topic, group in historical_data.groupby("topic"):
        correct = topic_accuracies.get(topic, {}).get("correct_answers", 0)
        total = topic_accuracies.get(topic, {}).get("total_questions", 1)
        topic_accuracy = (correct / total) * 100
        topic_accuracy_results.append((topic, topic_accuracy))
    return pd.DataFrame(topic_accuracy_results, columns=["Topic", "Accuracy"])

topic_accuracy_df = compute_topic_accuracy(historical_responses_df, mock_topic_accuracies)
print(topic_accuracy_df)
```

```python
def generate_recommendations(accuracy, weak_topics):
    recommendations = []
    accuracy_num = float(accuracy.strip('%')) if isinstance(accuracy, str) else float(accuracy)

    if accuracy_num < 50:
        recommendations.append("Focus on foundational topics to build confidence.")

    recommendations.extend([f"Practice more questions on {topic}." for topic in weak_topics[:3]])

    return recommendations

weak_topics = topic_accuracy_df.sort_values(by="Accuracy")["Topic"].tolist()
recommendations = generate_recommendations(
    quiz_submission_main["accuracy"].iloc[0], weak_topics
)

print("\n\nPersonalized Recommendations:")
for rec in recommendations:
    print(f"- {rec}")
```

```python
def create_study_plan(weak_topics):
    plan = {}
    for topic in weak_topics:
        plan[topic] = "Practice 5 questions daily"
    return plan

study_plan = create_study_plan(weak_topics)
print("\n\nSuggested Study Plan:")
for topic, task in study_plan.items():
    print(f"{topic}: {task}")
```

```python
# Define student persona
def define_persona(accuracy):
    if accuracy > 80:
        return "High Performer"
    elif accuracy > 50:
        return "Steady Learner"
    else:
        return "Beginner"
persona = define_persona(float(quiz_submission_main["accuracy"].iloc[0].strip('%')))
print(f"\n\nStudent Persona: {persona}")
```

```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/recommendations', methods=['GET'])
def get_recommendations():
    return jsonify({
        "accuracy": float(quiz_submission_main["accuracy"].iloc[0].strip('%')),
        "persona": persona,
        "recommendations": recommendations,
        "study_plan": study_plan
    })

if __name__ == '__main__':
    app.run(debug=True)
```
